```python
import networkx as nx
import konect2networkx as k2n

def draw_konect_ust(name): # internal name of the konect graph
  G = k2n.get(name, download = True, logging = False)
  tree = nx.random_spanning_tree(G, None)

  print("\nA gráf csúcsainak száma:", G.number_of_nodes())
  print("Eredeti gráf éleinek száma:", G.number_of_edges())
  print("A generált ust éleinek száma:", tree.number_of_edges())
  
  nx.draw(tree, with_labels=True)
```

```python
draw_konect_ust('brunson_southern-women')
```

```python
draw_konect_ust('moreno_kangaroo')
```
